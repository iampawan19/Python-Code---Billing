```python
import pandas as pd

# Function to read Excel file
def read_excel_file(file_path):
    try:
        df = pd.read_excel(file_path)
        return df
    except Exception as e:
        print(f"Error reading file {file_path}: {e}")
        return None

# Replace 'your_file.xlsx' and 'dummy_file.xlsx' with the actual paths to your Excel files
raw_file_path = 'C:/Users/pawan/Ponyup -Python/Laketown Check-in Upload_7th Nov..xlsx'
dummy_file_path = 'C:/Users/pawan/Ponyup -Python/PonyUp-CheckIn_dump.xlsx'

# Read Excel files
raw_df = read_excel_file(raw_file_path)
dummy_df = read_excel_file(dummy_file_path)

# Check if the DataFrames are successfully loaded
if raw_df is not None and dummy_df is not None:
    # Rename "Unnamed: 2" column to "Bill Number"
    raw_df = raw_df.rename(columns={"Unnamed: 2": "Bill Number"})

    # Pivot unique "Bill Number" from raw_df and set as "Bill No" column in dummy_df
    dummy_df["Bill No"] = raw_df["Bill Number"].unique()

    # Calculate the sum of "Bill Amount" for each "Bill No" in raw_df
    bill_amount_sum = raw_df.groupby("Bill Number")["Bill amount"].sum()

    # Map the sum of "Bill Amount" to the corresponding "Bill No" in dummy_df
    dummy_df["Bill Amount"] = dummy_df["Bill No"].map(bill_amount_sum)

# Map "Mobile No" column of dummy_df with "Mobile no" column of raw_df based on "Bill No"
bill_mobile_mapping = raw_df.drop_duplicates("Bill Number").set_index("Bill Number")["Mobile no"]
dummy_df["Mobile No"] = dummy_df["Bill No"].map(bill_mobile_mapping)

# Map "Name" column of dummy_df with "Name" column of raw_df based on "Bill No"
name_mapping = raw_df.drop_duplicates("Bill Number").set_index("Bill Number")["Name"]
dummy_df["Name"] = dummy_df["Bill No"].map(name_mapping)

# Map "Date" column of dummy_df with "Check in date" column of raw_df based on "Bill No"
date_mapping = raw_df.drop_duplicates("Bill Number").set_index("Bill Number")["Check in date"]
dummy_df["Date"] = dummy_df["Bill No"].map(date_mapping)

# Map "Time" column of dummy_df with "Check in time" column of raw_df based on "Bill No"
time_mapping = raw_df.drop_duplicates("Bill Number").set_index("Bill Number")["Check in time"]
dummy_df["Time"] = dummy_df["Bill No"].map(time_mapping)

# Fill "Country Code" column with "91"
dummy_df['Country Code'] = '91'

# Fill "Country Code" column with "91"
dummy_df['Points'] = '0'

# Convert "Date" column to "YYYY-MM-DD" format
dummy_df['Date'] = pd.to_datetime(dummy_df['Date']).dt.strftime('%Y-%m-%d')

# Convert "Time" column to "HH:MM:SS" format
dummy_df['Time'] = pd.to_datetime(dummy_df['Time']).dt.strftime('%H:%M:%S')

# Print the updated dummy_df
print(dummy_df)


# Save the updated_dummy_df DataFrame to a CSV file
dummy_df.to_csv('D:/laketown checkIn 3 dec.csv', index=False)

# Print a message indicating the CSV file has been saved
print("Updated dummy sheet saved to 'updated dummy_sheet.csv'") 






 
```
